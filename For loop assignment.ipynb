```python
#4.Develop a program that finds the factorial of a given number using a `for` loop.
```

```python
# Function to calculate factorial
def factorial(n):
    # Initialize result to 1 as the factorial of 0 is 1
    result = 1

    # Use a for loop to multiply numbers from 1 to n
    for i in range(1, n + 1):
        result *= i

    return result

# Input: Get a number from the user
num = int(input("Enter a number: "))

# Calculate and print the factorial
print(f"The factorial of {num} is: {factorial(num)}")
```

```python
#9.Write a program that calculates the average of a list of numbers using a `for` loop.
```

```python
def calculate_average(numbers):
    total = 0
    count = 0

    for num in numbers:
        total += num
        count += 1

    if count == 0:
        print("Cannot calculate average for an empty list.")
        return

    average = total / count
    return average

# Example usage:
numbers_list = [10, 20, 30, 40, 50]
result = calculate_average(numbers_list)
print(f"The average of the list {numbers_list} is: {result}")
```

```python
#10.Develop a program that prints the first `n` Fibonacci numbers using a `for` loop.
```

```python
def fibonacci_sequence(n):
    fib_numbers = []

    if n >= 1:
        fib_numbers.append(0)
    if n >= 2:
        fib_numbers.append(1)

    for i in range(2, n):
        fib_numbers.append(fib_numbers[i - 1] + fib_numbers[i - 2])

    return fib_numbers

# Example usage:
n = int(input("Enter the value of n for Fibonacci sequence: "))
fibonacci_numbers = fibonacci_sequence(n)

print(f"The first {n} Fibonacci numbers are:")
for num in fibonacci_numbers:
    print(num, end=" ")
```

```python
#11.Write a program to check if a given list contains any duplicates using a `for` loop.
```

```python
def has_duplicates(lst):
    seen = set()

    for element in lst:
        if element in seen:
            return True
        seen.add(element)

    return False

# Example usage:
my_list = [1, 2, 3, 4, 5, 2]  # This list contains duplicates (2)
result = has_duplicates(my_list)

if result:
    print("The list contains duplicates.")
else:
    print("The list does not contain duplicates.")
```

```python
#12.Create a program that prints the prime numbers in a given range using a `for` loop.
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def print_primes_in_range(start, end):
    if start < 2:
        start = 2

    print(f"Prime numbers in the range ({start}, {end}):")
    for number in range(start, end + 1):
        if is_prime(number):
            print(number, end=" ")

# Example usage:
start_range = int(input("Enter the start of the range: "))
end_range = int(input("Enter the end of the range: "))
print_primes_in_range(start_range, end_range)
```

```python
#14.Write a program to find the maximum element in a 2D list using a nested `for` loop.
```

```python
def find_max_element(matrix):
    if not matrix:
        return None  # Return None for an empty matrix

    max_element = matrix[0][0]

    for row in matrix:
        for element in row:
            if element > max_element:
                max_element = element

    return max_element

# Example usage:
matrix = [
    [1, 5, 3],
    [8, 2, 4],
    [6, 7, 9]
]

result = find_max_element(matrix)
print(f"The maximum element in the 2D list is: {result}")
```

```python
#17.Write a program that converts a list of Fahrenheit temperatures to Celsius using a `for` loop.
```

```python
def fahrenheit_to_celsius(fahrenheit_temps):
    celsius_temps = []

    for fahrenheit in fahrenheit_temps:
        celsius = (fahrenheit - 32) * 5.0/9.0
        celsius_temps.append(celsius)

    return celsius_temps

# Example usage:
fahrenheit_list = [32, 68, 104, 212]
celsius_result = fahrenheit_to_celsius(fahrenheit_list)

print("Fahrenheit temperatures:", fahrenheit_list)
print("Celsius temperatures:", celsius_result)
```

```python
#20.Write a program to find the greatest common divisor (GCD) of two numbers using a `for` loop.
```

```python
def find_gcd(num1, num2):
    smaller = min(num1, num2)

    for i in range(smaller, 0, -1):
        if num1 % i == 0 and num2 % i == 0:
            return i

# Example usage:
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))

gcd_result = find_gcd(number1, number2)

print(f"The GCD of {number1} and {number2} is: {gcd_result}")
```

```python
#22.Write a program to find the prime factors of a given number using a `for` loop and list comprehension.
```

```python
def find_prime_factors(number):
    factors = [factor for factor in range(2, number + 1) if (number % factor == 0 and all(factor % i != 0 for i in range(2, int(factor**0.5) + 1)))]

    return factors

# Example usage:
input_number = int(input("Enter a number to find its prime factors: "))
prime_factors_result = find_prime_factors(input_number)

if prime_factors_result:
    print(f"The prime factors of {input_number} are: {prime_factors_result}")
else:
    print(f"{input_number} has no prime factors.")
```

```python
#25.Write a program to flatten a nested list using list comprehension.
```

```python
def flatten_nested_list(nested_list):
    return [item for sublist in nested_list for item in (sublist if isinstance(sublist, list) else [sublist])]

# Example usage:
nested_list = [1, [2, 3, [4, 5]], 6, [7, 8]]
flattened_list_result = flatten_nested_list(nested_list)

print(f"Original nested list: {nested_list}")
print(f"Flattened list: {flattened_list_result}")
```

```python
#31.Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list comprehension.
```

```python
def sieve_of_eratosthenes(limit):
    primes = [True] * (limit + 1)
    primes[0] = primes[1] = False

    [primes.__setitem__(multiple, False) for num in range(2, int(limit**0.5) + 1) if primes[num] for multiple in range(num*num, limit + 1, num)]

    return [num for num in range(2, limit + 1) if primes[num]]

# Example usage:
limit = int(input("Enter the limit for prime numbers: "))
prime_numbers_result = sieve_of_eratosthenes(limit)

print(f"Prime numbers up to {limit}: {prime_numbers_result}")
```

```python
#32.Create a program that generates a list of all Pythagorean triplets up to a specified limit using list comprehension.
```

```python
def generate_pythagorean_triplets(limit):
    triplets = [(a, b, c) for a in range(1, limit + 1) for b in range(a, limit + 1) for c in range(b, limit + 1) if a**2 + b**2 == c**2]
    return triplets

# Example usage:
limit = int(input("Enter the limit for Pythagorean triplets: "))
pythagorean_triplets_result = generate_pythagorean_triplets(limit)

print(f"Pythagorean triplets up to {limit}: {pythagorean_triplets_result}")
```

```python
#34.Write a program that calculates the mean, median, and mode of a list of numbers using list comprehension.
```

```python
from statistics import mean, median, mode

def calculate_statistics(numbers):
    # Using list comprehension to calculate the mean
    mean_value = mean(numbers)

    # Median and mode don't lend themselves well to list comprehension
    # because they require sorted data and additional processing
    median_value = median(numbers)
    mode_value = mode(numbers)

    return mean_value, median_value, mode_value

# Example usage:
numbers_list = [1, 2, 3, 4, 5, 5, 6, 6, 7, 8, 9]
mean_result, median_result, mode_result = calculate_statistics(numbers_list)

print(f"Original list: {numbers_list}")
print(f"Mean: {mean_result}")
print(f"Median: {median_result}")
print(f"Mode: {mode_result}")
```

```python
#35.Create a program that generates Pascal's triangle up to a specified number of rows using list comprehension.
```

```python
from math import comb

def generate_pascals_triangle(rows):
    triangle = [[comb(n, k) for k in range(n + 1)] for n in range(rows)]
    return triangle

def display_triangle(triangle):
    for row in triangle:
        print(row)

# Example usage:
num_rows = int(input("Enter the number of rows for Pascal's Triangle: "))
pascals_triangle_result = generate_pascals_triangle(num_rows)

print(f"Pascal's Triangle with {num_rows} rows:")
display_triangle(pascals_triangle_result)
```

```python
#37.Write a program that finds the longest word in a sentence using list comprehension.
```

```python
def find_longest_word(sentence):
    # Split the sentence into words
    words = sentence.split()

    # Use list comprehension to find the length of each word
    word_lengths = [len(word) for word in words]

    # Find the index of the longest word
    max_length_index = word_lengths.index(max(word_lengths))

    # Return the longest word
    return words[max_length_index]

# Example usage:
input_sentence = "This is an example sentence to find the longest word"
result = find_longest_word(input_sentence)
print(f"The longest word in the sentence is: {result}")
```

```python
#40.Write a program that generates a list of prime palindromic numbers using list comprehension.
```

```python
def is_prime(num):
    # Function to check if a number is prime
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def is_palindrome(num):
    # Function to check if a number is a palindrome
    return str(num) == str(num)[::-1]

# Use list comprehension to generate a list of prime palindromic numbers
prime_palindromes = [num for num in range(1, 1000) if is_prime(num) and is_palindrome(num)]

print("Prime palindromic numbers:", prime_palindromes)
```
